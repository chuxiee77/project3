## Step 4

Use [`merge()`](https://pandas.pydata.org/docs/user_guide/merging.html#merge) to combine the two, and output the resulting table.

```python
population["Borough"] = population["Borough"].astype(str).str.strip()
project["Borough"] = project["Borough"].astype(str).str.strip()

merged = pd.merge(population,project,on = "Borough", how = "inner")
merged
```

## Step 5

Using the two datasets above, use pandas to produce an aggregate per-capita statistic by borough.

The dataset you chose before may not work for this. That's fine, pick another.

### Hint

You're creating a "number of [thing] per capita by borough" table.

1. Do a [`groupby()`](https://pandas.pydata.org/docs/user_guide/groupby.html) on the original dataset.
1. Join with the populations by borough.
1. Compute the per-capita values as a new column.

```python
project_counts = project.groupby('Borough').size()
project_counts
```

```python
project_counts_df = project_counts.to_frame(name = "Project Count")
project_counts_df
```

```python
project_counts_df = project_counts.reset_index(name = "Project Count")
project_counts_df
```

```python
merged_data = pd.merge(project_counts_df, population[['Borough', '2020']], on='Borough', how='inner')
merged_data['Granted Projects per Capita'] = merged_data['Project Count'] / merged_data['2020']
merged_data
```
